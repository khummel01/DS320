### real

```python
coursesDict = {'major':[],
        'course_title':[],
        'course_num':[],
        'hours':[],
        'geneds':[],
        'prereqs':[],
        'description':[]}

for major in dept_list:
    majorName = major.get('href','None')
    page = requests.get("http://www.luther.edu/catalog/curriculum/{}".format(majorName))
    soup = BeautifulSoup(page.text, 'html.parser')   
    
    for title in soup.find_all('span', class_='courseTitle'):
        coursesDict['course_title'].append(title.text)
        coursesDict['major'].append(majorName)
    for num in soup.find_all('span', class_='courseNumber'):
        coursesDict['course_num'].append(num.text)
    for courseDescription in soup.find_all('span', class_='courseDescription'):
        coursesDict['description'].append(courseDescription.text)
    for courseHrs in soup.select('ul.fa-ul'):
        coursesDict['hours'].append(courseHrs.text[1])

    fulfills = []
    fulfills_count = 0
    for e in soup.find_all('span',class_='genEd'):
        e = e.text
        fulfills.append(e)
    for x in soup.select('ul.fa-ul'):
        if "Fulfills:" not in x.text:
            coursesDict['geneds'].append('None')
        else:
            coursesDict['geneds'].append(fulfills[fulfills_count])
            fulfills_count += 1
            
    for x in soup.select('ul.fa-ul'):
        if "Prerequisites" in x.text:
            x= x.text
            x = x[23:]
            coursesDict['prereqs'].append(x)
        else:
            coursesDict['prereqs'].append('None')
```
